After running your data above through your machine learning model, you determine that a better prediction might come from breaking the personality type into its four parts (one part for each character in the type).  Your task is now to create a DataFrame called `Q16` that splits the personality type into the various parts and creates a new column for each part (these columns should be called `interaction` that will represent the first letter in the personality type, `information` for the second letter, `decision` for the third, and `organization` for the fourth).



```python
Q16_prep = Q15_prep['personality_type'].astype(str).apply(list)
split = pd.DataFrame(Q16_prep.tolist(), columns = ['interaction', 'information', 'decision', 'organization'])
Q16_prep2 = pd.concat([Q16_prep, split], axis =1).drop({'personality_type'}, axis =1 )
Q16_prep2['interaction'] = Q16_prep2['interaction'].map({'I': int(0), 'E': int(1)})
Q16_prep2['information'] = Q16_prep2['information'].map({'S': int(0), 'N': int(1)})
Q16_prep2['decision'] = Q16_prep2['decision'].map({'T': int(0), 'F': int(1)})
Q16_prep2['organization'] = Q16_prep2['organization'].map({'J': int(0), 'P': int(1)})
for i in ['interaction', 'information', 'decision', 'organization']:
    Q16_prep2[i] = Q16_prep2[i].astype('Int32')
Q16_prep = Q16_prep2.join(castaway_details)
Q16 = Q16_prep[['castaway_id', 'full_name', 'personality_type', 'interaction', 'information', 'decision', 'organization']].sort_values(by = 'castaway_id')
Q16
```

Using data from `castaways`, create a DataFrame called `Q17` that bins the contestant ages (their age when they were on the season, not their current age) into the following age categories:
- 18-24
- 25-34
- 35-44
- 45-54
- 55-64
- 65+



```python
Q17_prep = castaways.copy()

Q17_prep['age'] = Q17_prep['age'].astype(float)
bins = [17,24,34,44,54,64,np.inf]
Q17_prep.loc[:, 'age_category'] = pd.cut(Q17_prep['age'],bins, labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+'])
Q17_prep['age_category'] = Q17_prep['age_category'].cat.reorder_categories(['18-24', '25-34', '35-44', '45-54', '55-64', '65+'])
Q17_prep['age'] = Q17_prep['age'].astype(int)
Q17 = Q17_prep[['season', 'castaway_id', 'full_name', 'age', 'age_category']].sort_values(['age','castaway_id']).reset_index(drop= True)
```

 Based on the age categories you created above, what are the normalized percentages for the various age categories using `value_counts()`.  Sort the value counts by index.  Save this as `Q18`.

```python
Q18 = Q17.age_category.value_counts(normalize= True, sort = False)            
Q18
```
